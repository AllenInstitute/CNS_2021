# Transcriptomic data

Here, we have already downloaded the (large) transcriptomic expression data file. This data file can be found on [this mouse Patch-seq page](https://portal.brain-map.org/explore/classes/multimodal-characterization/multimodal-characterization-mouse-visual-cortex) under the link "Directory of Gene Expression Matrices" in the Downloads table. It is a file ending in `_cpm`, and the `v2` version of the file contains both the original cells released as well as additional data released later.

In this file, expression is quantified as counts-per-million (cpm), and each row is a gene and each column is a cell. 

```python
gene_data = pd.read_csv(
    "downloads/20200513_Mouse_PatchSeq_Release_cpm.v2.csv",
    index_col=0,
)
```

```python
gene_data
```

Let's create a 2D projection of the data using the transcriptomic data to see the structure of the different subclasses and types. We will transform the data, then use the UMAP algorithm to create a nonlinear embedding of a subset of the differentially expressed genes. 

The marker gene file (`select_markers.csv`) was calculated using the differential gene expression functions of the [`scrattch.hicat`](https://github.com/AllenInstitute/scrattch.hicat) R package using the large dissociated transcriptomic data set from the study by Tasic et al. (2018). This file can be found in this tutorial's GitHub directory.

```python
import umap

marker_genes_for_umap = pd.read_csv("select_markers.csv", index_col=0) # this file can be found in the GitHub repository for the tutorial

embedding = umap.UMAP(n_neighbors=25).fit_transform(
    np.log2(gene_data.loc[marker_genes_for_umap["Gene"], :].values.T + 1)
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(*embedding.T, s=1, edgecolor="none")
sns.despine()
```

We can first look at where the cells we found in the web explorer fall on this UMAP plot. The example downloaded specimen IDs file (`specimens-ids.csv`) can be found in this tutorial's GitHub repository.

```python
# The downloaded file includes the URL used to filter the cells, so we are skipping those lines
downloaded_ids = np.loadtxt("specimens-ids.csv", skiprows=4, dtype=int).tolist()

# get the transcriptomic sample IDs from the cell specimen IDs

downloaded_cells_sample_ids = metadata.loc[
    metadata["cell_specimen_id"].isin(downloaded_ids),
    "transcriptomics_sample_id"]
downloaded_cells_mask = gene_data.columns.isin(downloaded_cells_sample_ids.tolist())

plt.figure(figsize=(8, 8))
plt.scatter(*embedding.T, s=1, edgecolor="none")
plt.scatter(*embedding[downloaded_cells_mask, :].T, s=5, edgecolor="none")
sns.despine()
```

We can also see where specific types fall in this embedding by using the metadata to identify specific cells.

```python
# Identify the cells from a particular t-type (Lamp5 Plch2 Dock5)
my_ttype_metadata = metadata.loc[metadata["T-type Label"] == "Lamp5 Plch2 Dock5", :]

my_ttype_metadata
```

```python
my_ttype_mask = gene_data.columns.isin(my_ttype_metadata["transcriptomics_sample_id"].tolist())

plt.figure(figsize=(8, 8))
plt.scatter(*embedding.T, s=1, edgecolor="none")
plt.scatter(*embedding[my_ttype_mask, :].T, s=5, edgecolor="none")
sns.despine()
```

Finally, we can also look at how expression of a particular gene varies across types in this embedding.

```python
plt.figure(figsize=(8, 8))
plt.scatter(
    *embedding.T,
    s=1,
    c=np.log2(gene_data.loc["Npy", :].values + 1),
    vmin=0,
    cmap="viridis",
    edgecolor="none"
)
```

# Electrophysiology

Let's get the NWB file that has the electrophysiology data for this cell and use the [IPFX library](https://github.com/AllenInstitute/ipfx) to process it. We would ideally use the [DANDI archive's download utility](https://github.com/dandi/dandi-cli) to obtain the NWB file after obtaining its URL from the file manifest.

**Note**: The DANDI command-line tool is not currently working with the URLs in the file manifest, apparently due to changes at the DANDI archive. You can still access [the NWB files via the web](https://gui.dandiarchive.org/#/dandiset/000020), and you can use the information in the `file_name` column to help find an individual NWB file from the web. You can also still use the DANDI command line utility to download the entire data set at once.

```python
my_specimen_id = my_ttype_with_recon_metadata.iloc[0, :]["cell_specimen_id"]

nwb_urls = file_manifest.loc[
    (file_manifest["cell_specimen_id"] == float(my_specimen_id)) &
    (file_manifest["file_type"] == "nwb"),
    :
]
```

```python
nwb_path = nwb_urls["file_name"].values[0]
nwb_path

# Here I have already downloaded sub-643830424_ses-645169915_icephys.nwb
# from the web site instead of using the DANDI utility
```

```python
from ipfx.dataset.create import create_ephys_data_set
from ipfx.data_set_features import extract_data_set_features
from ipfx.utilities import drop_failed_sweeps

data_set = create_ephys_data_set(nwb_file=nwb_path)
drop_failed_sweeps(data_set)
cell_features, sweep_features, cell_record, sweep_records, _, _ = \
    extract_data_set_features(data_set, subthresh_min_amp=-100.0)
```

# Morphology

Now we'll get the SWC file that has the morphological reconstruction of this cell (using the FTP URL to get it from the Brain Image Library) and use the [neuron_morphology library](https://github.com/AllenInstitute/neuron_morphology) to process it.

```python
swc_urls = file_manifest.loc[
    (file_manifest["cell_specimen_id"] == float(my_specimen_id)) &
    (file_manifest["file_type"] == "transformed_swc"),
    :
]
```

```python
swc_urls["archive_uri"].values[0]
```

```python
swc_path = swc_urls["file_name"].values[0]
swc_path
```

```python
from neuron_morphology.swc_io import morphology_from_swc
from neuron_morphology.constants import AXON, BASAL_DENDRITE, SOMA
```

```python
morph = morphology_from_swc(swc_path)
```

We can make a quick plot of the reconstruction.

```python
axon_nodes = morph.get_node_by_types([AXON])
dend_nodes = morph.get_node_by_types([BASAL_DENDRITE])
soma_nodes = morph.get_node_by_types([SOMA])

axon_x = [node['x'] for node in axon_nodes]
axon_y = [node['y'] for node in axon_nodes]

dend_x = [node['x'] for node in dend_nodes]
dend_y = [node['y'] for node in dend_nodes]

soma_x = [node['x'] for node in soma_nodes]
soma_y = [node['y'] for node in soma_nodes]


plt.figure(figsize=(10, 10))

plt.scatter(axon_x, axon_y, s=1, edgecolor="none")
plt.scatter(dend_x, dend_y, s=1, edgecolor="none")
plt.scatter(soma_x, soma_y, s=20, c="black", edgecolor="none")


plt.gca().set(xticks=[], yticks=[])
plt.gca().set_aspect("equal")
sns.despine(left=True, bottom=True)
```

Then we can calculate a set of morphological features.

```python
from neuron_morphology.feature_extractor.data import Data
from neuron_morphology.feature_extractor.feature_extractor import FeatureExtractor
from neuron_morphology.features.dimension import dimension
from neuron_morphology.feature_extractor.marked_feature import nested_specialize
from neuron_morphology.feature_extractor.feature_specialization import NEURITE_SPECIALIZATIONS
from neuron_morphology.features.statistics.coordinates import COORD_TYPE_SPECIALIZATIONS


features_to_calculate = [
    nested_specialize(
        dimension,
        [COORD_TYPE_SPECIALIZATIONS, NEURITE_SPECIALIZATIONS]),
]

fe = FeatureExtractor()
fe.register_features(features_to_calculate)
feature_extraction_run = fe.extract(Data(morph))
```

```python
feature_extraction_run.results
```

```python
width = feature_extraction_run.results["axon.node.dimension"]["width"]
height = feature_extraction_run.results["axon.node.dimension"]["height"]
wh_ratio = width / height
print(width, height, wh_ratio)
```
